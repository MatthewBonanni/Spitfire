```python
from spitfire.chemistry.mechanism import ChemicalMechanismSpec
from spitfire.chemistry.flamelet import Flamelet
import numpy as np
from time import perf_counter
import h5py as h5io
from datetime import datetime
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

m = ChemicalMechanismSpec(cantera_xml='nHeptane.LiuHewsonChenPitsch04_highT.xml', group_name='gas')

pressure = 101325.

air = m.stream('TPY', (298., pressure, 'O2:0.232, N2:0.778'))
fuel = m.stream('TPY', (485., pressure, 'NXC7H16:1'))
z_stoich = m.stoich_mixture_fraction(fuel, air)
nzi = 128

x_range = np.logspace(-2, np.log10(160), 30)

flamelet_specs = {'mech_spec': m,
                  'pressure': pressure,
                  'oxy_stream': air,
                  'fuel_stream': fuel,
                  'grid_points': nzi + 2,
                  'grid_cluster_intensity': 4.,
                  'grid_cluster_point': 0.03,
                  'initial_condition': 'equilibrium',
                  'max_dissipation_rate': x_range[0]}

flamelet = Flamelet(**flamelet_specs)
```

```python
plt.plot(flamelet.mixfrac_grid, flamelet.initial_temperature, '.-')
plt.xlabel('mixture fraction')
plt.ylabel('temperature (K)')
plt.title('equilibrium state')
plt.grid(True)
plt.show()
```

```python
for s in ['O2', 'NXC7H16']:
    plt.semilogy(flamelet.mixfrac_grid, flamelet.initial_mass_fraction(s), '-', label=s)
plt.legend(loc='best')
plt.xlabel('mixture fraction')
plt.ylabel('mass fractions')
plt.title('equilibrium state')
plt.grid(True)
plt.show()
```

```python
# specifications for the solver
solve_specs = {'steady_tolerance': 1.e-8,
               'log_rate': 10,
               'write_log': False,
               'transient_tolerance': 1.e-8,
               'max_time_step': 1.e4,
               'first_time_step': 1.e-4}

# how the HDF5 files will be written out
filename_template = '/sierra/dev/mahanse/spitfire-tabprops-test/spitfire-attempt/heptane_for_john_chi{:}'

# quantities to tabulate
tabulated_quantities = ['temperature', 'density', 'mass fraction OH', 'viscosity']

# loop over dissipation rates
running_state = None
for idx, x in enumerate(x_range):
    # set the dissipation rate and use the prior solution as initial guess
    flamelet_specs['initial_condition'] = running_state if running_state is not None else 'equilibrium'
    flamelet_specs['max_dissipation_rate'] = x
    
    # helpful for the first few dissipation rates
    solve_specs['first_time_step'] = 1.e-4 if idx < 2 else 1.e-2

    # make a flamelet and solve with the transient-to-steady integrator (tends to work best for lower stream temperatures)
    flamelet = Flamelet(**flamelet_specs)
    cput0 = perf_counter()
    flamelet.integrate_to_steady(**solve_specs)
    dcput = perf_counter() - cput0
    running_state = flamelet.final_state
    print('converged flamelet {:4}/{:4} (chi_max = {:8.1e} 1/s) in ({:7.2f} s), T_max = {:6.1f}'
          .format(idx + 1, x_range.size, x, dcput, np.max(running_state)))

    # use scipy's interpolant to get the stoichiometric dissipation rate
    x_stoich = interp1d(flamelet.mixfrac_grid, flamelet._x)(z_stoich)
    
    # compute quantities for tabulation
    for k in tabulated_quantities:
        if k == 'viscosity':
            flamelet.insitu_process_cantera_method('viscosity')
        else:
            flamelet.insitu_process_quantity(k)
    data_dict = flamelet.process_quantities_on_state(flamelet.state_with_bcs(running_state))

    # write the HDF5 database for this particular flamelet
    with h5io.File(filename_template.format(str(idx).zfill(3)), 'w') as hf:
        flamelet_group = hf.create_group('flamelet')
        flamelet_group.create_dataset('mixture_fraction', data=flamelet.mixfrac_grid)
        for q in tabulated_quantities:
            if q == 'mass fraction OH':
                flamelet_group.create_dataset('mass_fraction_OH', data=data_dict[q].ravel())
            flamelet_group.create_dataset(q, data=data_dict[q].ravel())
        flamelet_group.attrs['mechanism'] = np.string_('nHeptane.LiuHewsonChenPitsch04_highT')
        flamelet_group.attrs['fuel'] = np.string_('n-heptane')
        flamelet_group.attrs['author'] = np.string_('Mike Hansen, Sandia National Laboratories mahanse@sandia.gov')
        flamelet_group.attrs['date'] = np.string_(str(datetime.now()))
        flamelet_group.attrs['title'] = np.string_('n-heptane-air case for John Hewson')
        flamelet_group.attrs['pressure'] = np.string_(str(pressure * 1.01325 / 101325) + '[bar]')
        flamelet_group.attrs['dissipation_rate_stoich'] = np.string_(str(x_stoich) + '[1/s]')
        flamelet_group.attrs['dissipation_rate_max'] = np.string_(str(x) + '[1/s]')
        flamelet_group.attrs['z_stoich'] = np.string_(z_stoich)
        flamelet_group.attrs['T_max'] = np.string_(str(np.max(flamelet.final_temperature)) + '[K]')
        flamelet_group.attrs['air_T'] = np.string_(str(air.T) + '[K]')
        flamelet_group.attrs['fuel_T'] = np.string_(str(fuel.T) + '[K]')
        for idx, s in enumerate(m.gas.species_names):
            Y_air = air.Y[idx]
            Y_fuel = fuel.Y[idx]
            if Y_air > 1.e-14:
                flamelet_group.attrs['air_mass_fraction_' + s] = np.string_(str(Y_air))
            if Y_fuel > 1.e-14:
                flamelet_group.attrs['fuel_mass_fraction_' + s] = np.string_(str(Y_fuel))
```
